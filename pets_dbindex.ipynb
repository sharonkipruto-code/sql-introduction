```python
import sqlite3
import pandas  as pd 

conn=sqlite3.connect('data.sqlite')
cur=conn.cursor()
```

```python
#order by clause- used to sort items in ascending/descending order
pd.read_sql('''SELECT * FROM products
            ORDER BY productName;''',conn).head()
```

```python
pd.read_sql('''SELECT * FROM products
            ORDER BY productName DESC;''',conn).head()
```

```python
pd.read_sql('''SELECT productVendor,productName, MSRP FROM products
            ORDER BY productName,productVendor ;''',conn).head()
```

```python
pd.read_sql('''SELECT productName,LENGTH(productDescription) AS description_length
             FROM products
            ORDER BY description_length DESC ;''',conn).head()
```

```python
pd.read_sql('''SELECT COUNT(DISTINCT productVendor) AS num_product_vendors,
            COUNT(DISTINCT productName) AS num_product_names
            FROM products;''',conn)
```

```python
pd.read_sql('''SELECT comments
            FROM ORDERS ORDER BY LENGTH(comments );''',conn).head()
```

## Groupby/having

```python
pd.read_sql('''SELECT country , COUNT(*) AS customer_count
            FROM customers 
            GROUP BY country;''',conn).head(10)

# returns the number of customers per country
```

```python
pd.read_sql('''SELECT customerNumber , 
            COUNT(*) AS number_payment,
            MIN(amount) AS min_purchase,
            MAX(amount) AS max_purchase,
            AVG(amount) AS AVG_purchase,
            SUM(amount) AS total_spent
            FROM payments 
            GROUP BY customerNumber;''',conn)
```

```python
pd.read_sql('''SELECT customerNumber , 
            COUNT(*) AS number_payment,
            MIN(amount) AS min_purchase,
            MAX(amount) AS max_purchase,
            AVG(amount) AS AVG_purchase,
            SUM(amount) AS total_spent
         FROM payments 
        WHERE strftime('%Y',paymentDate)='2004'
        GROUP BY customerNumber;''',conn)
```

```python
pd.read_sql('''SELECT customerNumber , 
            COUNT(*) AS number_payment,
            MIN(amount) AS min_purchase,
            MAX(amount) AS max_purchase,
            AVG(amount) AS AVG_purchase,
            SUM(amount) AS total_spent
         FROM payments 
        GROUP BY customerNumber
        HAVING AVG(amount)>50000;''',conn)
```

```python
pd.read_sql('''SELECT customerNumber,
            COUNT(*) AS number_payment,
            MIN(amount) AS min_purchase,
            MAX(amount) AS max_purchase,
            AVG(amount) AS AVG_purchase,
            SUM(amount) AS total_spent
        FROM payments
        WHERE amount > 50000
        GROUP BY customerNumber
        HAVING COUNT(*) >= 2
        ORDER BY total_spent
        LIMIT 1;''', conn).head(10)
```
