```python
#concatenate strings with || lets add a period after initial 
pd.read_sql("""
SELECT substr(firstName, 1, 1) || "." AS first_initial
  FROM employees;
""", conn).head()
```

```python
import sqlite3
import pandas as pd
conn=sqlite3.connect('data.sqlite')
```

```python
cur=conn.cursor()
```

```python
pd.read_sql('''SELECT lastName,firstName email FROM employees
            WHERE lastname ="Patterson";''',conn)
```

```python
pd.read_sql("""
            SELECT lastName,firstName, email 
            FROM employees
            WHERE length(firstName)=5;
            """, conn)
```

```python
pd.read_sql("""
            SELECT *, strftime('%m', orderDate) AS month 
            FROM orders
            WHERE month="01";
            """, conn)
```
